### 1. Основи роботи з умовними конструкціями (if-elif-else)

```python
cooking_time = 5
cooking_time
```

```python
if cooking_time < 4:
    print("Яйце рідке")
elif 4 <= cooking_time < 7:
    print("Яйце всмятку")
elif 7 <= cooking_time < 10:
    print("Яйце вкруту")
else:
    print("Яйце переварене")
```

### 2. Перевірка оцінки

```python
grade = 72
grade
```

```python
if grade >= 90:
    print("Відмінно")
elif 75 <= grade < 90:
    print("Добре")
elif 60 <= grade < 75:
    print("Задовільно")
else:
    print("Незадовільно")
```

### 3. Перевірка на наявність

```python
list_goods = ["молоко", "яйця", "масло", "сир"]
print(list_goods)
```

```python
find_eggs = "яйця"
find_bread = "хліб"
if find_eggs in list_goods:
    print("Яйця є в холодильнику")
else:
    print("Яєць немає")
```

```python
result = "Хліб є в списку" if find_bread in list_goods else "Хліба нема"
print(result)
```
