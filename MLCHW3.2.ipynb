### 5. Статистика із групуванням

Обчисліть (кожен пункт оцінюється в 0.5 бала):

```python
#Середній рахунок за статтю (groupby sex)
```

```python
by_sex = df.groupby('sex')['total_bill'].mean()
print("Середній чек за статтю:")
print(by_sex)
```

```python
#Середні чайові за днем тижня (groupby day)
```

```python
by_day = df.groupby('day')['tip'].mean()
print("Середні чайові за днем тиждня:")
print(by_day)
```

```python
#Скільки відвідувачів у кожен день
```

```python
week_or = ['Thur', 'Fri', 'Sat', 'Sun']
df['day'] = pd.Categorical(df['day'], categories=week_or, ordered=True)

visit_by_day = df.groupby('day', observed=False)['size'].sum()
print("Кількість відвідувачів за днем тиждня:")
print(visit_by_day)
```

### 6. Створення нової колонки

Створіть нову колонку tip_percentage - відсоток чайових від загальної суми рахунку. Виведіть перші 10 рядків з новою колонкою.

```python
df['tip_percentage'] = ((df['tip'] / df['total_bill']) * 100).round(2).map("{} %".format)
print(df['tip_percentage'].head(10))
```
