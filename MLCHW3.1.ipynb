### 2. Базові операції

```python
# СтСтворіть Numpy-масив зі значеннями [12, 24, 36, 48, 60].

# Виконайте наступні операції над оригінальним масивом і запишіть у нову змінну результат 
# (кожен пункт оцінюється в 0.5 бала):

# Помножте всі елементи на 2
# Додайте до всіх елементів 10
# Піднесіть всі елементи до квадрату
# Обчисліть квадратний корінь з усіх елементів
# Виведіть результат виконання кожної операції.
```

```python
arr2 = np.array ([12, 24, 36, 48, 60])
print(arr2) 
```

```python
arr_mult = 2 * arr2
print(arr_mult)
```

```python
arr_plus = 10 + arr2
print(arr_plus)
```

```python
arr_deg = arr2 ** 2
print(arr_deg)
```

```python
arr_roo = np.sqrt(arr2) 
print(arr_roo)
```

```python
print(arr2) 
print(arr_mult)
print(arr_plus)
print(arr_deg)
print(arr_roo)
```

### 3. Статистичні функції

```python
# Створіть Numpy-масив data зі значеннями [18, 27, 33, 41, 52, 68, 79].

# Обчисліть використовуючи функціонал numpy (кожен пункт оцінюється в 0.5 бала):

# Середнє значення
# Мінімум
# Максимум
# Суму всіх елементів
```

### 4. Індексація

```python
# Створіть Numpy-масив arr зі значеннями [15, 22, 35, 44, 58, 67].

# Виведіть (кожен пункт оцінюється в 0.5 бала):

# Перший елемент
# Останній елемент
# Елементи з індексами 1 до 3
# Елементи більші за 40
```

```python
arr = np.array([15, 22, 35, 44, 57, 67])
print(arr)
```

```python
print(arr[0])
```

```python
print(arr[-1])
```

```python
# Елементи з індексами 1 до 3
print(arr[1:3])
```

```python
# Елементи з індексами 1 до 3 включно
print(arr[1:4])
```

```python
over_40 = arr[arr > 40]
print(over_40)
```
