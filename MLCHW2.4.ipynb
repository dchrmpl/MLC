### 1. Цикл for - базовий

```python
#Використовуючи цикл for, виведіть числа від 1 до 7 включно використовуючи функцію range.
#Варіант 1

for n in range(8):
    if n > 0:
        print(n)
```

```python
#Варіант 2

for n in range(1, 8):
    print(n)
```

### 2. Цикл for по списку

```python
#Дано список:

#names = ["Олег", "Анна", "Іван", "Марія"]
#Використовуючи цикл, виведіть кожне ім'я з привітанням: "Привіт, [ім'я]!"

names = ["Олег", "Анна", "Іван", "Марія"]

for name in names:
    print(f"Привіт, {name}!")
```

### 3. Цикл з enumerate

```python
#Дано список оцінок:
#grades = [85, 92, 78, 95, 88]
#Використовуючи enumerate, виведіть кожну оцінку з номером: "Студент 1: 85 балів"

grades = [85, 92, 78, 95, 88]
for m, grade in enumerate(grades):
    print(f"Студент {m+1}: {grade} балів")
```

### 4. Сума парних чисел

```python
#Напишіть програму, яка обчислює суму всіх парних чисел від 1 до 20 включно.

sum_p = list(range(1, 21))
print (sum_p)
res = 0
for s_p in sum_p:
    if s_p % 2 == 0:
        res = res + s_p
print(res)
```
