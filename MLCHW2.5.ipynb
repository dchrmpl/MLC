### 1. Цикл for - базовий

```python
#Задано змінну counter, яка рівна -3.
#Напишіть програму, використовуючи while, яка в циклі додає до змінної counter +1 і зупиняється, 
#коли змінна counter стане суворо більше за 5. 
#На кожній ітерації циклу виводьте значення змінної counter.
```

```python
counter = -3
while counter < 5:
    counter = counter + 1
    print(counter)
```

### 2. List comprehension - прості операції

```python
#Дано список чисел:
#numbers = [11, 4, 9, 16, 7, 5]
#Створіть нові списки використовуючи list comprehension, які містять (кожен пункт оцінюється в 0.5 бала):
#Квадрати всіх чисел
#Тільки парні числа
#Всі числа помножені на 10
#Виведіть новоутворені списки.
```

```python
numbers = [11, 4, 9, 16, 7, 5]
numbers_sq = [n**2 for n in numbers]
print(numbers_sq)
```

```python
numbers = [11, 4, 9, 16, 7, 5]
numbers_sqp = [n**2 for n in numbers if n % 2 == 0]
print(numbers_sqp)
```

```python
numbers = [11, 4, 9, 16, 7, 5]
numbers_d = [n*10 for n in numbers]
print(numbers_d)
```

```python
print(numbers_sq, numbers_sqp, numbers_d, sep='\n')
```

### 3. List comprehension - комбінована логіка

```python
#Дано список чисел від 1 до 10. Створіть новий список, де виконується наступна логіка:
#Якщо число парне - помножте його на 2
#Якщо непарне - віднімніть 1
#Виведіть результуючий список на екран.
```

```python
numbers = list(range(1, 11))
print(numbers)
numbers_upg = [n*2 if n % 2 == 0 else n-1 for n in numbers]
print(numbers_upg)
```
