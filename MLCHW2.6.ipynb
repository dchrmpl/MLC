### 1. Функція з параметром

```python
#Створіть функцію show_menu(), яка приймає перелік страв, наприклад ['Піца', 'Паста', 'Салат'] 
#і виводить меню ресторану в наступному вигляді:

#=== МЕНЮ ===
#1. Піца
#2. Паста
#3. Салат
#=============
#Викличте цю функцію на списку ["Суп грибний", "Котлета по-київськи", "Лимонад"].
```

```python
def show_menu(dish):
    print("=== МЕНЮ ===")
    for numb, d in enumerate(dish, start=1):
        print(f"{numb}. {d}")
    print("=============")
dish = ['Піца', 'Паста', 'Салат']
show_menu(dish)
```

```python
dish = ["Суп грибний", "Котлета по-київськи", "Лимонад"]
show_menu(dish)
```

### 2. Функція з параметром

```python
#Створіть функцію print_square(n), яка виводить квадрат числа.

#Наприклад: print_square(5) → "Квадрат числа 5 = 25"

#Викличте її з різними числами.
```

```python
def print_square(n):
    result = n ** 2
    print(f"Квадрат числа {n} = {result}")
```

```python
print_square(10)
print_square(27)
print_square(2)
```

### 3. Функція з поверненням значення

```python
#Створіть функцію multiply(a, b), яка повертає добуток двох чисел.

#Викличте функцію на власноруч заданих значеннях та виведіть результат.
```

### 4. Функція пошуку максимумального значення

```python
#Створіть функцію find_max(numbers), яка:

#Приймає список чисел
#Повертає найбільше число в списку
#Протестуйте на списку [15, 42, 8, 23, 56, 19].
```

```python
def find_max(numbers):
    return max(numbers)
```

```python
list_exp = [15, 42, 8, 23, 56, 19]
max = find_max(list_exp)
print(max)
```

### 5. Функція з умовами

```python
#Створіть функцію check_grade(grade), яка:

#Приймає оцінку (число)
#Повертає "Здав" якщо grade >= 60
#Повертає "Не здав" якщо grade < 60
#Протестуйте на різних оцінках.
```

```python
def check_grade(grade):
    if grade >= 60:
        return "Здав"
    else:
        return "Не здав"
```

```python
grade_1 = 15
result = check_grade(grade_1)
print(result)
```

```python
grade_2 = 80
result = check_grade(grade_2)
print(result)
```

```python
grade_3 = 56
result = check_grade(grade_3)
print(result)
```

```python
grade_4 = 190
result = check_grade(grade_4)
print(result)
```
